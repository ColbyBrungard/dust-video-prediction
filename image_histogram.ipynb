```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import color
from skimage.transform import rescale
```

# RGB Dust Image and Histogram 

```python
Image1 = '/Users/josue/Documents/Work/Video_Project/Results/Images_Thesis/Original/Image_1/dust_2134.png' # path to image

C = 0.1 # scaling coefficient 

I1 = mpimg.imread(Image1); # read in image
I1 = rescale(I1,C) # scale image 

plt.figure(figsize=(25,5)) # figure size
# plot image
plt.subplot(1,4,1)
plt.imshow(I1)
plt.axis('off')
plt.title('Dust Image')
# plot band 1 histogram
plt.subplot(1,4,2)
plt.hist((I1[:,:,0]*255).flatten(),bins=255,range=(0,255));
plt.title('Red Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 2 histogram
plt.subplot(1,4,3)
plt.hist((I1[:,:,1]*255).flatten(),bins=255,range=(0,255));
plt.title('Green Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 3 histogram
plt.subplot(1,4,4)
plt.hist((I1[:,:,2]*255).flatten(),bins=255,range=(0,255));
plt.title('Blue Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.tight_layout()
# plt.savefig('/Users/josue/Desktop/Scaled_RGB_dust_Histogram.png', format='png', dpi=1200) # save file
```


# RGB Overlap Histograms

```python
Image1 = '/Users/josue/Documents/Work/Video_Project/Results/Images_Thesis/Original/Image_1/dust_2134.png' # path to image
Image2 = '/Users/josue/Documents/Work/Video_Project/Results/Images_Thesis/Original/Image_6/non-dust_2134.png' # path to image

I1 = mpimg.imread(Image1); # read in image
I2 = mpimg.imread(Image2); # read in image

C = 0.1 # scaling coefficient 

I1 = rescale(I1,C) # scale image 
I2 = rescale(I2,C) # scale image 

plt.rcParams.update({'font.size': 60})
plt.figure(figsize=(50,10)) # figure size
# plot band 1 histogram
plt.subplot(1,3,1)
plt.hist((I1[:,:,0]*255).flatten(),bins=255,range=(0,255));
plt.hist((I2[:,:,0]*255).flatten(),bins=255,range=(0,255));
plt.title('Red Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 2 histogram
plt.subplot(1,3,2)
plt.hist((I1[:,:,1]*255).flatten(),bins=255,range=(0,255));
plt.hist((I2[:,:,1]*255).flatten(),bins=255,range=(0,255));
plt.title('Green Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 3 histogram
plt.subplot(1,3,3)
plt.hist((I1[:,:,2]*255).flatten(),bins=255,range=(0,255));
plt.hist((I2[:,:,2]*255).flatten(),bins=255,range=(0,255));
plt.title('Blue Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.tight_layout()
# plt.savefig('/Users/josue/Desktop/Scaled_RGB_Overlap_Histogram.png', format='png', dpi=1200) # save file
# plt.savefig('/Users/josue/Desktop/RGB_Overlap_Histogram.png', format='png', dpi=1200) # save file
```


# HSV Overlap Histogram 

```python
Image1 = '/Users/josue/Documents/Work/Video_Project/Results/Images_Thesis/Original/Image_1/dust_2134.png' # path to image
Image2 = '/Users/josue/Documents/Work/Video_Project/Results/Images_Thesis/Original/Image_6/non-dust_2134.png' # path to image

I1 = color.rgb2hsv(mpimg.imread(Image1)); # read in image and convert rbg to hsv
I2 = color.rgb2hsv(mpimg.imread(Image2)); # read in image and convert rbg to hsv

C = 0.1 # scaling coefficient 

I1 = rescale(I1,C) # scale image 
I2 = rescale(I2,C) # scale image 


plt.rcParams.update({'font.size': 60})
plt.figure(figsize=(50,10)) # figure size
# plot band 1 histogram
plt.subplot(1,3,1)
plt.hist((I1[:,:,0]*255).flatten(),bins=255,range=(0,255), label = 'Dust Frame');
plt.hist((I2[:,:,0]*255).flatten(),bins=255,range=(0,255), label = 'Non Dust Frame');
plt.title('Hue Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 2 histogram
plt.subplot(1,3,2)
plt.hist((I1[:,:,1]*255).flatten(),bins=255,range=(0,255), label = 'Dust Frame');
plt.hist((I2[:,:,1]*255).flatten(),bins=255,range=(0,255), label = 'Non Dust Frame');
plt.title('Saturation Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
# plot band 3 histogram
plt.subplot(1,3,3)
plt.hist((I1[:,:,2]*255).flatten(),bins=255,range=(0,255), label = 'Dust Frame');
plt.hist((I2[:,:,2]*255).flatten(),bins=255,range=(0,255), label = 'Non Dust Frame');
plt.title('Value Band')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.tight_layout()
# plt.savefig('/Users/josue/Desktop/Scaled_HSV_Overlap_Histogram.png', format='png', dpi=1200) # save file
```
